## Меры изменчивости

**Размах (range)** - разность между max и min значениями. 
**Дисперсия (variance)** - средний квадрат отклонений индивидуальных значений признака от из средней величины. Среднеквадратическое отклонение - квадратный корень из дисперсии. sd - для выборки, сигма - для генеральной совокурности.  
*Формула дисперсии для ГС:*  
$$\frac{\displaystyle\sum_{i=1}^{n}(x_i - \mu)^2} {n}$$  
*Формула дисперсии для выборки:*  
$$\frac{\displaystyle\sum_{i=1}^{n}(x_i - \mu)^2} {n-1}$$   

**Свойства дисперсии и sd**  
$$D_(x+c)=D_x$$
$$sd_(x+c)=sd_x$$

$$D_(xc)=D_x*c^2$$
$$sd(xc)=sd_x*c$$

## ЦПТ

https://gallery.shinyapps.io/CLT_mean/

Признак имеент нормальное распределение. Многократно извлекаем выборки одинакового объема, строим распределение выборочных средних. Такое распределение будет являться нормальным, среднее совпадает с ГС со стандартным отклонением (стандартная ошибка среднего)=сигма ГС(или sd выборки)/корень из числа наблюдений. Чем больше наблюдений, тем меньше изменчивость признака.  
Стандартная ошибка всегда меньше, чем стандартное отклонение  
- Распределение выборочных средних является нормальным, со средним равным среднему значению признака в генеральной совокупности  
- Стандартная ошибка среднего - это среднеквадратическое отклонение распределения выборочных средних  
- Чем меньше стандартная ошибка среднего, тем реже выборочные средние будут сильно отклоняться от среднего в генеральной совокупности  
- Стандартная ошибка среднего тем меньше, чем больше объем выборки и меньше вариативность исследуемого признака

## Доверительные интервалы для среднего

95% доверительный интервал: центр - выборочной среднее, слева выборочное среднее - 1,96 * se, справа + 1,96 * se.   
99% доверительный интервал: центр - выборочной среднее, слева выборочное среднее - 2,58 * se, справа + 2,58 * se. 

- Мы можем быть на 95% уверены, что среднее значение в генеральной совокупности принадлежит рассчитанному доверительному интервалу.
- Если многократно повторять эксперимент, для каждой выборки рассчитывать свой доверительный интервал, то в 95 % случаев истинное среднее будет находиться внутри доверительного интервала.

## T-распределение

Распределение Стьюдента - если число наблюдений невелико и станд отклонение неизвестно (t-distribution).  
Наблюдения с большецй вероятностью попадают за пределы +- 2сигма.  
Число степеней свободы. df=n-1. С увеличением числа df распределение стремится к нормальному.  


Правильнее будет сказать, что мы используем t - распределение не потому что у нас маленькие выборки, а потому что мы не знаем стандартное отклонение в генеральной совокупности. Поэтому в дальнейшем мы всегда будем использовать t - распределение для проверки гипотез, если нам неизвестно стандартное отклонение в генеральной совокупности, необходимое для расчета стандартной ошибки, даже если объем выборки больше 30.  
t-критерий рассчитывается также как z.

## Сравнение двух средних; t-критерий Стьюдента

Первая выборка - это выборка из пациентов, которые не принимали новое лекарство. А вторая выборка - из пациентов, которые его принимали.

Таким образом, если мы сравниваем две выборки по t-критерию:

нулевая гипотеза - лекарство не работает, выборки принадлежат одной ГС

альтернативная гипотеза - лекарство работает, выборки принадлежат разным ГС

$$t=\frac{\displaystyle\overline X_1 - \overline X_2 } {\sqrt{\frac{sd_1^2} {n_1} + \frac{sd_2^2} {n_2}}}$$ 

$$df=n_1+n_2-2$$

Гомогенность дисперсии (дисперсии примерно одинаковы) можно оценить, используя "Критерий Левена" или "Критерий Фишера"

## Однофакторный дисперсионный анализ

Очень часто в экспериментах и исследованиях возникает необходимость сравнить несколько групп между собой. В таком случае мы можем применять однофакторный дисперсионный анализ.  Та переменная, которая будет разделять наших испытуемых или наблюдения на группы (номинативная переменная с нескольким градациями) называется независимой переменной. А та количественная переменная, по степени выраженности которой мы сравниваем группы, называется зависимая переменная. 

1) Вычисляем среднее значение **всех** наблюдений.  
2) SST - общая сумма квадратов (изменчивость без учета разделения на группы)  
3) df=N-1  
Общая сумма квадратов может проистекать из 2-х источников: SSB (междугрупповая сумма) и SSW (внутригрупповая сумма, показатель дисперсии в каждой группе).  
сумма по всем группам $$SSW=\sum (X_i-\overline X)^2$$
$$df=N-m$$ где m=кол-во групп  
сумма по всем группам $$SSB=\sum m*(\overline X - \overline{\overline{X}})^2$$
$$df=m-1$$  
Если большая часть SST за счет SSB - значит общая изменчивость за счет изменчивости между группами.  
$$F=\frac {\frac {SSB} {m-1}} {\frac {SSW} {N-m}}$$
распределение Фишера F-значение принимает только положительные значения
